```python
d = {'a' : 1, 'b' : 2, 'c' : 3}
ser = pd.Series(data = d)  # 딕셔너리의 키가 인덱스로 생성됨
ser
```

```python
# 시리즈를 접근하는 방법 : 시리즈[인덱스]
print(ser.index, ser.values, ser.value_counts())
print()
print(ser_list.index, ser_list.values)
print()

# ser의 'b' 인덱스 값을 선택
print("ser['b']의 값:", ser['b'], ser[1])  # 시리스[인덱스명] 또는 시리즈(숫자 인덱스)
print()
# 여러 개의 값을 선택 : 시리즈[['a', 'c']]
print(ser[['a','c']])  # 인덱스가 여러개면 리스트 안에 나열
print()
# 범위를 선택할 때 : 시리즈[start: stpp], 인덱스명으로 검색할 경우는 stop에 포함, 숫자 인덱스로 검색할 경우는 stop 포함 X
print("index명 ", ser['a':'b'])
print()
print('숫자 인덱스 : ', ser[0:1])
```

```python
list_a = list(range(5))
sr = pd.Series(list_a, index = ['a','b','c','d','e'])
print(sr)
#   문 : 처음부터 마지막까지 자료를 출력하는데 2씩 증가시키면서
# 인덱스명을 사용
print(sr['a':'e':2])
# 정수형 인덱스를 사용
print(sr[0:len(sr):2])  # 시리즈 [start : stop : step]
```

```python
# 데이터프레임 : 시리즈가 모여서 2차원 배열의 형태를 가짐
dict_a = {'c0' : [1,2,3], 'c1' : [4,5,6], 'c2' : [7,8,9]}
print(dict_a)
print()

# dict를 데이터프레임으로 변경
df = pd.DataFrame(dict_a)  # 딕셔너리의 키가 컬럼 인덱스가 됨
print(df)
print()

# dict를 데이터프레임으로 변경, 인덱스명을 부여하면서
df = pd.DataFrame(dict_a, index = ['a', 'b', 'c'])
print(df)
print()

print(df.index)
print(df.columns)
```

```python
# 데이터프레임의 인덱스명 재부여 : df.index = [새로운 인덱스명]
df.index = ['k1', 'k2', 'k3']
print(df)
print()

# 데이터프레임의 컬럼명 재부여 : df.columns = [새로운 컬럼명]
df.index = ['c3', 'c4', 'c5']
print(df)
print()
```

```python
# 2차원 리스트를 데이터프레임으로
# pd.DataFrame(data, index=[], columns=[])
list_a = [[1,2,3], [4,5,6], [7,8,9]]
df = pd.DataFrame(list_a, index = ['a', 'b', 'c'], columns = ['c0', 'c1', 'c2'])
print(df)
```

```python
# 인덱스명 -> a,b,c  컬럼명 -> c0, c1, c2로 변경
df.index = ['a', 'b', 'c']
df.columns = ['c0', 'c1', 'c2']
print(df)

# 데이터프레임에서 c0 컬럼의 값만 추출
print(df[['c0']]); print()
# 데이터프레임에서 c0와 c2컬럼의 값만 추출
print(df[['c0', 'c2']]); print()

# c1에서 c2까지 추출 -> 행과 열을 같이 나열 : df.loc[행 인덱스명, 컬럼 인덱스명]
# 행을 인덱스명으로 추출 -> df.loc(행 인덱스명, 컬럼 인덱스명)
# 행을 정수형 인덱스로 추출 -> df.iloc(행 정수형 인덱스, 컬럼 정수형 인덱스)
print(df.loc[:, 'c1':'c2']); print()
df.iloc[:, 1:3] # [:] -> 전체 자료 (처음부터 끝까지) 출력하는 것을 의미

# df에서 1행과 2행, c1 컬럼과 c2 컬럼 추출
print(df.loc[['a','b'], ['c1', 'c2']]); print()
df.iloc[:-1, :-1]

print(df.iloc[0, 0:1])
```

```python
# 시리즈 연산 : 시리즈 사칙연산(+,-,*,/) 숫자 -> 시리즈의 각 원소에 숫자를 사칙연산한 결과 반환
sr1 = pd.Series({'국어':100, '영어':90, '수학':95})
print(sr1 + 100)
print(sr1 / 20)
print(sr1 - 30)
```

```python
# 시리즈 연산 : 시리즈 사칙연산(+,-,*,/) 시리즈 -> 순서대로 계산 X, 먼저 인덱스를 기준으로 정렬한 후 사칙연산 실행
sr2 = pd.Series({'수학':80, '국어':90, '영어':100})
sr1 + sr2 
```

```python
# 두 시리즈의 인덱스가 다른 경우 : 사칙연산 -> NaN 인 값과 더하게 되는 경우 계산 처리 안되고 그대로 NaN 값 처리
import numpy as np
sr1 = pd.Series({'국어': np.nan, '영어':90, '수학':95})
sr2 = pd.Series({'수학':80, '국어':90})
sr1 + sr2
```

```python
# add(), sub(), mul(), div() 메서드 활용해서 NaN 값을 0으로 처리하면서 연산 실행
sr1.add(sr2, fill_value = 0)
sr1.sub(sr2, fill_value = 0)
```
