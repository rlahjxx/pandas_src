```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# "시도별 전출입 인구수.xlsx"
df = pd.read_excel('./dataset/시도별 전출입 인구수.xlsx')
df
```

```python
# 전출지가 '서울특별시'인 자료만 확인
df[df['전출지별'] == '서울특별시']
```

```python
#  전출지별에 있는 NaN 값 해결 : df.fillna(값, method = 'ffill')
                    # -> method : 결측값을 변경할 방식 (ffiil : 결측값을 바로 위의 값과 동일하게 변경, bfill : 바로 아래의 값으로 변경)
df.fillna(method='ffill', inplace = True)
df[df['전출지별'] == '서울특별시']
```

```python
# 전출지별이 서울특별시 이고, 전입지별이 서울특별시가 아닌 자료 추출 (서울에서 다른 지역으로 이동한 데이터만 추출하여 정리)
df_seoul = df[(df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')]
df_seoul
```

```python
# 전출지별이 모두 같은 값이므로 전출지별 컬럼 삭제
df_seoul = df_seoul.drop('전출지별', axis=1)
df_seoul
```

```python
# 전입지별을 전입지로 이름 변경
df_seoul.rename(columns = {'전입지별':'전입지'}, inplace=True)
df_seoul
```

```python
# 전입지 컬럼을 인덱스로 지정
df_seoul.set_index('전입지', inplace=True)
df_seoul
```

```python
# 서울에서 경기도로 전입한 인구 분석
df_one = df_seoul.loc['경기도']
df_one
```

```python
plt.plot(df_one.index, df_one.values) # x,y축 데이터를 plot 함수에 입력
```

```python
import matplotlib as mlt

mlt.rc('font', family = 'Malgun Gothic')  #  'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False  #  -기호 사용 가능  #  이부분 안하고 아랫부분하면 한글때문에 깨져서나옴
```

```python
# 서울에서 경기도로 이동한 인구의 변화를 라인 그래프로 그리기
plt.figure(figsize=(10,5))
plt.style.use('fivethirtyeight')

plt.plot(df_one, color='green', marker='^')

#   title
plt.title('서울 -> 경기도 이동 인구수')
plt.legend(labels = ['서울 -> 경기'])
plt.xlabel('연 도')
plt.xticks(rotation=75)
plt.ylabel('인구 수')

# y축 범위 지정 (최소값, 최대값)
plt.ylim(50000, 800000)

# 주석 표시 - 화살표
plt.annotate('',
             xy=(20, 620000),       #화살표의 머리 부분(끝점)
             xytext=(2, 290000),    #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='r', lw=3), #화살표 서식
             )

plt.annotate('',
             xy=(47, 450000),       #화살표의 머리 부분(끝점)
             xytext=(30, 580000),   #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='olive', lw=5),  #화살표 서식
             )

# 주석 표시 - 텍스트
plt.annotate('인구이동 증가(1970-1995)',  #텍스트 입력
             xy=(10, 380000),            #텍스트 위치 기준점
             rotation=30,                #텍스트 회전각도
             va='baseline',              #텍스트 상하 정렬
             ha='center',                #텍스트 좌우 정렬
             fontsize=10,                #텍스트 크기
             )

plt.annotate('인구이동 감소(1995-2017)',  #텍스트 입력
             xy=(40, 500000),            #텍스트 위치 기준점
             rotation=-12,               #텍스트 회전각도
             va='baseline',              #텍스트 상하 정렬
             ha='center',                #텍스트 좌우 정렬
             fontsize=10,                #텍스트 크기
             )

plt.show()
```

```python
# 경기도 인구 이동의 그래프를 하나의 페이지에 2개를 1열에 1개 2행 1열의 그래프 그림
# fig.add_subplot(행의 갯수, 열의 갯수, 순번)
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(df_one, color = 'r', label = '서울-> 경기')
ax1.legend()
ax1.set_title('서울-> 경기')
ax2.plot(df_one, color = 'b')
ax2.set_title('서울-> 경기')

ax1.set_xticklabels(df_one.index, rotation=75)  # subplot에서는 xticks가 아니라 xticklabels라고 해야 rotation 가능
# z축 눈금 라벨 크기
ax1.tick_params(axis='x', labelsize=10)
ax1.tick_params(axis='y', labelsize=15)
ax2.set_xticklabels(df_one.index, rotation=75)
plt.show()
```

```python
# 서울시에서 충남, 경북, 강원도로 이동한 인구수의 그래프를 그리고자 함
# 하나의 그래프에 중복해서 그리고자 함
df = pd.read_excel('./dataset/시도별 전출입 인구수.xlsx')
df = df.fillna(method='ffill')
df_seoul = df[(df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')]
df_seoul = df_seoul.drop(columns = ['전출지별'])
```

```python
df_seoul = df_seoul.set_index('전입지별')
df_seoul
```

```python
df_three = df_seoul.loc[['충청남도', '경상북도', '강원도']]
df_three
```

```python
# 그래프 그리기
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(1,1,1) # figure에 서브 플롯 생성
ax.plot(df_three.columns, df_three.loc['충청남도', :], color='blue', label = '서울->충남', marker = 'o', lw=2) # lw -> 라인 굵기 설정
ax.plot(df_three.columns, df_three.loc['경상북도', :], color='green', label = '서울->경북', marker = '.', lw=2)
ax.plot(df_three.columns, df_three.loc['강원도', :], color='orange', label = '서울->강원', marker = '^', lw=2)

ax.legend()
ax.set_title('서울 -> 충남, 경북, 강원 이동 인구', size=15)
ax.set_xlabel('기간', size=10)
ax.set_ylabel('이동 인구 수', size=10)

ax.set_xticklabels(df_three.columns, rotation=75, size=10)

plt.show()
```

```python
# 전라북도 자료를 추가해서 df_four를 생성
df_four = df_three.copy()
df_four.loc['전라북도'] = df_seoul.loc['전라북도', :]
df_four
```

```python
# df_four 그래프 그리기
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(2,2,1) # 충남
ax2 = fig.add_subplot(2,2,2) # 경북
ax3 = fig.add_subplot(2,2,3) # 강원
ax4 = fig.add_subplot(2,2,4) # 전북

col_years = df_four.columns
ax1.plot(col_years, df_four.loc['충청남도', :], color = 'r', label = '서울->충남', marker = 'o', lw=2)
ax2.plot(col_years, df_four.loc['경상북도', :], color = 'g', label = '서울->경북', marker = 'o', lw=2)
ax3.plot(col_years, df_four.loc['강원도', :], color = 'b', label = '서울->강원', marker = 'o', lw=2)
ax4.plot(col_years, df_four.loc['전라북도', :], color = 'y', label = '서울->전북', marker = 'o', lw=2)

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()

ax1.set_title('서울 -> 충남 이동 인구 수')
ax2.set_title('서울 -> 경북 이동 인구 수')
ax3.set_title('서울 -> 강원 이동 인구 수')
ax4.set_title('서울 -> 전북 이동 인구 수')

plt.show()
```

```python
# pandas plot() 함수로 면적 그래프 작성
# x축에 눈금 -> 정수형이어야 하기 때문에 transpose 함수를 통해 전치행렬로 변환 필요
df_four = df_four.T
df_four
```

```python
df_four.index = df_four.index.map(int)
df_four.index
```

```python
# 면적 그래프 <- 인덱스가 x축으로, 컬럼별로 면적 그래프 그림
df_four.plot(kind='area', stacked = False, alpha = 0.2, figsize = (20, 10)) # stacked = True 기본 설정값, alpha : 투명도(기본 설정값 1)

plt.title('서울 -> 충남, 경북, 강원, 전북의 이동 인구 수')
plt.xlabel('기간')
plt.ylabel('이동 인구 수')
plt.show()
```

```python
# 면적 그래프 <- 인덱스가 x축으로, 컬럼별로 면적 그래프 그림
ax = df_four.plot(kind='area', stacked = False, alpha = 0.2, figsize = (20, 10)) # stacked = True 기본 설정값, alpha : 투명도(기본 설정값 1)

ax.set_title('서울 -> 충남, 경북, 강원, 전북의 이동 인구 수')
ax.set_xlabel('기간')
ax.set_ylabel('이동 인구 수')
ax.legend()
plt.show()
```
