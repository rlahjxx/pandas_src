```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('./dataset/air_final.csv')
df2 = pd.read_csv('./dataset/traffic_final.csv')
```

# 시간대 및 통행량 컬럼 재생성

```python
s = df2[['일자', '주소명']]
v = df2.loc[:, '24시':'23시']
df2 = pd.concat([s,v], axis=1)
```

```python
df2
```

```python
df_group = df2.groupby('일자')
r = pd.DataFrame()

for key, group in df_group:
    for i in range(1,25):
        s = group[['일자', '주소명', str(i)+'시']].copy()
        s['시간'] = str(i) + '시'
        s.rename(columns = {str(i)+'시':'차량수'}, inplace=True)
        r = pd.concat([r,s])
```

```python
r['주소명'].value_counts()
r
```

```python
df_end = pd.merge(df1, r, left_on = ['air_date', 'air_hour', 'air_place'], right_on = ['일자', '시간', '주소명'])
```

```python
df_end['air_place'].value_counts()
df_end
```

```python
# 겹치는 컬럼 제거 및 차량수 컬럼 데이터 타입 변경
df_end.drop(['일자', '주소명', '시간'], axis=1, inplace=True)
df_end = df_end.astype({'차량수':'int'})
df_end
```

```python
# 통행량 컬럼 이름 변경
df_end.rename(columns = {'차량수':'air_car'}, inplace=True)
df_end['air_place'].value_counts()
```

```python
# 개수가 맞지 않는 강동구의 행 제거
df_final = df_end[df_end['air_place']!='강동구']
```
