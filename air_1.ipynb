```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./dataset/air_2.csv')
```

# 타입 변경

```python
# 일자를 데이터타입으로 변경
df = df.astype({'air_date':'str'})
df['air_date'] = pd.to_datetime(df['air_date'])
df
```

```python
df.info()
```

```python
# 결측값 전처리
df.replace({'-':0}, inplace=True)
df
```

```python
df[['air_pm1', 'air_pm2']] = df[['air_pm1', 'air_pm2']].astype(int)
```

```python
df.loc[:, 'air_Ozone':'air_so'] = df.loc[:, 'air_Ozone':'air_so'].astype(float)
df.info()
```

# join을 위한 시간 컬럼값 통일

```python
df['air_hour'] = df['air_hour'].str.split(':').str[0]
df
```

```python
df['air_hour'] = df['air_hour'].astype(int)
df['air_hour'] = df['air_hour'].astype(str) + '시'
df
```
