```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('dataset/traffic.xlsx', sheet_name = [2])
df
```

```python
df = pd.DataFrame(df[2])
df
```

```python
# 밑에 있는 설명 2행 잘라내기
df = df[0:139]
df
```

```python
# 필요없는 행 잘라내기
df2 = df.copy()
df2.drop(['검지기 유형', '도로명 주소', '유입 방향', '유출방향', '지점번호'], axis=1, inplace=True)
df2
```

```python
# 주소에서 자치구만 추출헌 컬럼 생성 후 기존 주소명 컬럼 삭제
df2['주소명'] = df2['주소'].str.split(' ').str[1]
df2
```

```python
s = df2[df2['주소명'].str.contains('시')]
s
```

```python
df2.drop(s.index, inplace=True)
df2
```

```python
df3 = df2.copy()
df3.drop_duplicates(subset=['주소명'], inplace=True)
df3.reset_index(drop=True, inplace=True)
```

```python
df3
```

```python
df3.drop([9,17], inplace=True)
df3
```

```python
df3.set_index('주소명', inplace=True)
df3
```

```python
df3.drop(['지점명칭', '주소'], axis=1, inplace=True)
df3
```
