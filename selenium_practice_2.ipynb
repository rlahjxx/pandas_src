```python
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys  # 폴더 경로 안에 있는 Keys 모듈 사용
import time  # 시간 변수 사용
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By


# 크롬 드라이버 사용을 위해 크롬 버전 확인 필요
# (...에 도움말 - Chrome 정보 - 버전 확인 후 버전에 맞는 크롬 드라이버 다운로드)
driver_path = 'C:/Users/enduser/Python Basic/semi_project/chromedriver.exe'
driver = webdriver.Chrome(driver_path)
```

```python
url = 'https://cleanair.seoul.go.kr/airquality/localAvg'
driver.get(url)
```

```python
# 찾고자 하는 데이터의 id나 class를 먼저 찾는다(복사한 id나 class에 공백이 있는 경우 지우고 .으로 바꿔줘야 함)
# id는 보통 하나만 쓰지만, class의 경우 여러번 사용할 수 있기 때문에 element -> elemnets로 먼저 찍어본다
```

```python
list_a = []
s = '2022-07-'
for i in range(1,8):
    if i < 10:
        num = '0' + str(i)  # '{:02d}'.format(i) 도 사용 가능
    else :
        num = str(i)
    s1 = s+ num
#     print(s1)   
    driver.find_element_by_id('searchDate').clear()
    driver.find_element_by_id('searchDate').send_keys(s1)
    time.sleep(2)
    for j in range(1,25):
        driver.find_element_by_id('searchTime')
        select = Select(driver.find_element_by_id('searchTime'))
        select.select_by_index(j)
        time.sleep(2)
        driver.find_element_by_id('search').click()
        
        while True:
            try:
                table = driver.find_element_by_id('localAvg-table')
                tbody = table.find_element_by_tag_name('tbody')
                tr = tbody.find_elements_by_tag_name('tr')
                print(len(tr)
                for k in tr:
                    print(k.text.split())
                    list_a.append(k.text.split())              
                time.sleep(3)
                break
            except:
                print('오류로 재 실행')               
df = pd.DataFrame(list_a)
df.drop_duplicates(inplace=True)
```

```python
df.columns = ['air_date','air_hour','air_place','air_pm1','air_pm2','air_Ozone','air_no','air_co','air_so']
df
```

```python
df.to_csv('C:/Users/enduser/Python Basic/semi_project/dataset/air_1.csv', index=None, encoding='utf-8-sig')
```
