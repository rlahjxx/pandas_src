# 그룹 연산

```python
# 1단계 -> 기준이 되는 컬럼으로 분할
# 2단계 -> 그룹에 메서드 적용
# 3단계 -> 결과를 결합
# titanic에서 age, class, sex, fare, survived 컬럼만 가져와서 데이터프레임 생성
import seaborn as sns

df = sns.load_dataset('titanic').loc[:, ['age', 'class', 'sex', 'fare', 'survived']]
df
```

```python
df['class'].unique()  # class 컬럼으로 분할 -> df.groupby('컬럼명')
```

```python
grouped = df.groupby('class')  # class의 값을 기준으로 그룹을 생성 (숫자인 경우에만 사용 가능)
```

```python
for key, group in grouped:  # key에는 First, Second, Third
    print(f'key : {key} 의 number {len(group)}')
    print(group.head(3))
```

```python
grouped.mean()
grouped.sum()
grouped.median()
```

```python
# 그룹에서 하나의 그룹만 가져오기 : grouped.get_group('Third')
grouped.get_group('Third').head(3)
```

```python
# 2개 이상의 컬럼으로 그룹 : groupby([컬럼 리스트])
grouped_two = df.groupby(['class', 'sex'])
grouped_two.head(5)
```

```python
for key, group in grouped_two:
    print(f'key : {key}의 number {len(group)}')
    print(group.head(3))
```

```python
grouped_two.mean()
grouped_two.std()
```

```python
# df를 class 컬럼으로 그룹 생성
# 각 그룹에 대해서 함수 정의 : lambda x: x.max() - x.min() -> 그룹객체.agg(함수)
# 각 그룹에 대해 min(), max(), mean() 함수 구함
# 각 그룹의 'fare' 컬럼은 최소금액과 평균금액을
# 'age' 컬럼은 중간값을 구하세요
```

```python
# 더미변수를 이용하여 성별 숫자형으로 변환
df2 = pd.get_dummies (df.loc[:, df.columns != 'class'], drop_first = True)
df2 = pd.concat([df2, df['class']], axis=1)
df2
```

```python
# df를 class 컬럼으로 그룹 생성
grouped_1 = df2.groupby('class')
group_func1 = grouped_1.agg(lambda x: x.max() - x.min())
print(group_func1)
print()
group_func2 = grouped_1.agg(['min', 'max', 'mean'])  # 여러 함수를 사용하고 싶을 때 사용
print(group_func2)
print()
# 'fare' 컬럼은 최소금액과 평균금액을, 'age' 컬럼은 중간값
group_func3 = grouped_1.agg({'fare' : ['min', 'mean'], 'age' : ['median']})
group_func3
```

```python
# 그룹 연산 데이터 변환
age_mean = grouped_1.age.mean()  # 그룹별 나이 평균
print(age_mean['Second'])

age_std = grouped_1.age.std()  # 그룹별 나이 표준편차
age_std.First
```

```python
# 각 그룹의 각 행에서 나이 - 자신이 속한 그룹의 평균/표준편차를 구함
grouped = df2.groupby('class')
age_mean = grouped.age.mean()
age_std = grouped.age.std()
```

```python
for key, group in grouped.age:
    score = group - age_mean[key] / age_std[key]
    print(f'key : {key}')
    print(score.head())   
```

```python
# 그룹에 transform()을 이용하여 그룹의 각 행에 대해 그룹연산 수행 가능
age_score = grouped.age.transform(lambda x: (x - x.mean()) / x.std())
age_score[[1,9,0]]
```

```python
# 그룹 객체 필터링 : 그룹개체.filter(조건식 함수)
# 각 그룹의 데이터 개수가 200개 이상인 그룹만 선택
df_4 = grouped.filter(lambda x : len(x) >= 200)
df_4['class'].unique()
```

```python
# 그룹 객체에 함수 매핑 : group객체.apply(함수)
grouped.apply(lambda x: x.head())
```

```python
age_score_apply = grouped.age.apply(lambda x: (x - x.mean()) / x.std())
age_score_apply[[1,9,0]]
```

```python
# class와 sex로 그룹을 생성
grouped = df.groupby(['class', 'sex'])
# 각 그룹별 평균을 출력
print(grouped.mean())
print()

# age열의 평균 값이 30보다 작은 그룹을 출력
age_filter = grouped.age.apply(lambda x: x.mean() < 30)
age_filter

for index in age_filter.index:
    if age_filter.loc[index] == True:
        print(index)
print()

# age 열의 평균 값이 30보다 작은 그룹의 데이터를 출력하는데 처음 5개씩만
for key,group in grouped.age:
    if age_filter.loc[key] == True:
        print('key :  ', key)
        print(group.head())
```

```python
# 피벗 : 엑셀의 피벗과 유사
# 필요 요소 : 그룹의 기준이 되는 index, 서브 컬럼, 그룹 함수에 적용할 값
df = sns.load_dataset('titanic')[['class', 'sex', 'age', 'fare', 'survived']]
df
```

```python
pdf = pd.pivot_table(df,  # 사용할 데이터
                    index = ['class', 'sex'],   # 그룹으로 분리할 컬럼
                    columns = 'survived', # 컬럼의 위치에 들어갈 컬럼
                    values = ['age', 'fare'],  # 데이터로 사용할 값
                    aggfunc = ['mean', 'max']  # 데이터에 적용할 집계 함수
                    )
pdf
```

```python
pdf.xs('female', level='sex')
pdf.xs('female', level=1)

pdf.xs('mean', axis=1)
pdf.xs(('mean', 'age'), axis=1)
pdf.xs(('max', 'fare', 1), axis=1)
```
